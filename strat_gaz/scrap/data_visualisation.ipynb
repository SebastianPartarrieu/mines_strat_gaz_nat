```python
import os 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

# Data Vizualization Process
-----------------------

Data scrapped from EEX Website. We reached spot and future prices.

- Url = https://www.powernext.com/futures-market-data
- Url = https://www.powernext.com/spot-market-data

Please find following, a quick incite of all the data we collected for now.

```python
def list_csv_dir( directory, price_type ):
    files = list(filter(lambda x : x[-4:]=='.csv' and price_type in x,os.listdir(directory)))
    return files 
```

## Spot Prices

```python
price_type = 'spot'
files = list_csv_dir('./last_save/',price_type)
print(files)
plt.figure(figsize = (15,10))
for file in files:
    table = pd.read_csv('./last_save/'+file)
    table['Trading Day'] = pd.to_datetime(table['Trading Day'],format = '%Y-%m-%d')
    plt.plot(table['Trading Day'], table['Price'],label = file[:-4])
plt.legend()
min_date  = min(table['Trading Day'])
max_date = max(table['Trading Day'])
plt.title(f' depuis le {min_date.day} jusque le {max_date}')
```

## Prix Forward

```python
price_type = 'forward'
files = list_csv_dir('./last_save/',price_type)
plt.figure(figsize = (15,10))
for file in files:
    table = pd.read_csv('./last_save/'+file)
    table['Trading Day'] = pd.to_datetime(table['Trading Day'],format = '%Y-%m-%d')
    plt.plot(table['Trading Day'], table['Month+1'],label = file[:-4])
plt.legend()
min_date  = min(table['Trading Day'])
max_date = max(table['Trading Day'])
plt.title(f' depuis le {min_date.day} jusque le {max_date}')
```
